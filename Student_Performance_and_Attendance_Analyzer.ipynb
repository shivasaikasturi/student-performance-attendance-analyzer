```python
plt.figure(figsize=(8,5))

sns.histplot(df['GPA'], bins=10, kde=True)

plt.title("Distribution of GPA")
plt.xlabel("GPA")
plt.ylabel("Number of Students")

plt.show()
```

## 12. Student Participation (Raised Hands) vs GPA

```python
# Convert Raised Hands to numeric
df['Raised Hands'] = pd.to_numeric(df['Raised Hands'], errors='coerce')

# Check how many invalid values were found
print(df['Raised Hands'].isnull().sum())
```

```python
df['Raised Hands'].fillna(df['Raised Hands'].median(), inplace=True)
```

```python
print(df['Raised Hands'].dtype)
```

```python
plt.figure(figsize=(8,5))

sns.scatterplot(x='Raised Hands', y='GPA', data=df)

plt.title("Raised Hands vs GPA")

plt.xlabel("Raised Hands")
plt.ylabel("GPA")

plt.show()
```

```python
print(df[['Raised Hands',
          'Visited Resource',
          'Announcements Visited',
          'Discussion']].dtypes)
```

```python
cols = ['Raised Hands',
        'Visited Resource',
        'Announcements Visited',
        'Discussion']

for col in cols:
    df[col] = pd.to_numeric(df[col], errors='coerce')
    df[col] = df[col].fillna(df[col].median())
```

```python
plt.figure(figsize=(8,5))

sns.scatterplot(x='Visited Resource', y='GPA', data=df)

plt.title("Visited Resource vs GPA")
plt.xlabel("Visited Resource")
plt.ylabel("GPA")

plt.show()
```

### Insight

The scatter plot indicates that students who access learning resources more frequently generally tend to have higher GPA values. This suggests that active use of academic resources contributes positively to student performance.

```python
plt.figure(figsize=(8,5))

sns.scatterplot(x='Discussion', y='GPA', data=df)

plt.title("Discussion vs GPA")
plt.xlabel("Discussion")
plt.ylabel("GPA")

plt.show()
```

### Insight

Students who participate more actively in classroom discussions generally achieve better academic performance. The graph demonstrates a positive association between discussion participation and GPA.

```python
# Select only numeric columns
numeric_df = df.select_dtypes(include=['number'])

plt.figure(figsize=(10,8))

sns.heatmap(
    numeric_df.corr(),
    annot=True,
    cmap='coolwarm',
    linewidths=0.5
)

plt.title("Correlation Heatmap")

plt.show()
```

### Insight

The heatmap illustrates the correlation among numerical variables. Strong positive correlations indicate that variables increase together, while weak or negative correlations indicate little or inverse relationships. It helps identify the factors most strongly associated with academic performance.
